```python
def hasNumbers(inputString):
                return bool(re.search(r'\d', inputString))

def tender_data_v1(path, file):
    page_l=[]
    text_l=[]
    file_l=[]
    doc = fitz.open(os.path.join(path, file))
    for j in range(doc.pageCount):
      if j>30:
        page = doc.loadPage(j)
        text = ' '.join(page.getText().splitlines()).strip()
        text = text.splitlines()
        text1 = ' '.join([i.strip() for i in text if i != ''])
        text1= text1.replace('\s+', '')
        if ('audit' in text1.lower().strip()) and ('tender' in text1.lower().strip()) and ('plan'  in text1.lower().strip()):
            print(j)
            ind1 = [a.start() for a in list(re.finditer('audit', text1.lower()))]
            if ind1:
                for i in range(len(ind1)):
                    if ('tender' in text1[ind1[i]-200:ind1[i]+300].lower()) and hasNumbers(text1[ind1[i]-200:ind1[i]+300].lower()):
                        text_l.append(text1[ind1[i]-200:ind1[i]+300])
                        file_l.append(file)
                        page_l.append(j)
                
            
    df_c= pd.DataFrame()
    df_c['file'] = file_l
    df_c['page']  = page_l
    df_c['text'] = text_l
    
    df_c = df_c.drop_duplicates()
    return(df_c)
        
```

```python
df_tender = pd.DataFrame()

for i in os.listdir(path):
    print(i)
    a =tender_data_v1(path, i)
    df_tender = pd.concat([df_tender, a], axis = 0)
    
```

```python
df = pd.read_csv(r'D:\C Drive Backup 2021\Desktop\company_contracts\answers\tender_data\last_auditor\spacy_labels_v1.csv')
```

```python
def spacy_form(a):
    c={}
    b=[]
    a = eval(str(a))
    for i in a:
        start= i['start']
        end = i['end']
        
        label=''.join(i['labels'])
        # if label == 'PER':
        #   label = 'PERSON'
        # elif label == 'AREA':
        #   label= 'WORK_OF_ART'
       
        # elif label == 'TITLE':
        #   label = 'NORP'
        b.append((start, end, label))
    
    c['entities']=b
    return(c)

df= df.dropna()
df['spacy_format'] = df.label.apply(lambda x:spacy_form(x) )
df['train_v1'] =df[['text', 'spacy_format']].apply(lambda x: (x['text'], x['spacy_format']), axis = 1)
spacy_train = [i for i in df.train_v1]
```

```python
import spacy
nlp=spacy.blank("en")

nlp.add_pipe(nlp.create_pipe('ner'))
ner = nlp.get_pipe('ner')
for _, annotations in spacy_train:
    for ent in annotations.get('entities'):
        ner.add_label(ent[2])
```

```python
# spacy2.2.4
n_iter =80
import random
other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
with nlp.disable_pipes(*other_pipes):
    optimizer = nlp.begin_training()
    for i in range(n_iter):
        random.shuffle(spacy_train)
        losses = {}
        
        for texts,annotations in spacy_train:
            
            nlp.update([texts], [annotations],  sgd = optimizer, losses= losses, drop=0.2)
        print(i, losses)
            
```

```python
def combine_entities_auditor(data_group):
    
    dt = data_group.groupby(['file'])['entities'].apply(lambda x: sum(x, [])).reset_index(drop = False)
    
    return(dt)
```

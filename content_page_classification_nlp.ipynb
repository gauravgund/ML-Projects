# naive bayes

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['clean_text'], data['target'], random_state=1, test_size= 0.1)
```

```python
cv = CountVectorizer(strip_accents='ascii', token_pattern=u'(?ui)\\b\\w*[a-z]+\\w*\\b', lowercase=True, stop_words='english')
X_train_cv = cv.fit_transform(X_train)
X_test_cv = cv.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train_cv, y_train)
predictions = naive_bayes.predict(X_test_cv)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
print('Accuracy score: ', accuracy_score(y_test, predictions))
print('Precision score: ', precision_score(y_test, predictions))
print('Recall score: ' , recall_score(y_test, predictions))
```

```python
val = pd.DataFrame()
val['actual'] = list(y_test)
val['pred'] = list(predictions)
val['text'] = list(X_test)
```

# extract 260 reports for unseen data: cgt_data 2020

```python
import os
import pandas as pd
import fitz
import pickle as pkl


def clean_func(data):
    data['text'] = data.text.apply(lambda x: str(x))
    data['clean_text'] = data.text.apply(lambda x: x.replace('.', '').replace(',', '').replace(';', '').replace('%', '').replace(':', '').replace('$', ''))
    data['clean_text'] = data.clean_text.str.replace('\d+', '')
    data['clean_text'] = data.clean_text.str.encode('ascii', 'ignore').str.decode('ascii')
    data['clean_text']= data.clean_text.str.strip().replace('  ', '')
    return(data)

def page_classification_nlp_bulk(path, cv1, load_model, outfile):
    l_text=[]
    page_text=[]
    for i in range(len(os.listdir(path))):
        try:
            with fitz.open(os.path.join(path, os.listdir(path)[i])) as f:
                print(os.listdir(path)[i])
                for j in range(f.pageCount):
                    if j <= 6: 
                        page = f.loadPage(j)
                        r_text=''
                        blocks = page.getText("blocks")
                        blocks.sort(key=lambda block: block[1])  # sort vertically ascending
                        for b in blocks:
                            text = b[4].strip().splitlines()
                            if text:
                                if not text[0].startswith('<image:'):
                                    r_text += ' '.join(text)
                                else:
                                    continue 
                        l_text.append(r_text)
                        page_text.append(os.listdir(path)[i]+'_page_'+ str(j)+'.pdf')
        except:
            continue
    df = pd.DataFrame()
    df['text'] = l_text
    df['page']= page_text
    df1 = clean_func(df)
    x_unseen = cv1.transform(df1.clean_text)
    prediction_unseen = load_model.predict(x_unseen, raw_score=True)
    df1['class_pred'] = list(load_model.predict(x_unseen))
    df1['raw_score'] = list(prediction_unseen)
    df1.to_csv(outfile, index= False)
    print('DONE')
    
```

```python
def page_classification_nlp_single(file, cv1, load_model):
    l_text=[]
    page_text=[]
    try:
            with fitz.open(file) as f:
                
                for j in range(f.pageCount):
                    if j <= 6: 
                        page = f.loadPage(j)
                        r_text=''
                        blocks = page.getText("blocks")
                        blocks.sort(key=lambda block: block[1])  # sort vertically ascending
                        for b in blocks:
                            text = b[4].strip().splitlines()
                            if text:
                                if not text[0].startswith('<image:'):
                                    r_text += ' '.join(text)
                                else:
                                    continue 
                        l_text.append(r_text)
                        page_text.append(file+'_page_'+ str(j)+'.pdf')
    except:
            continue
    df = pd.DataFrame()
    df['text'] = l_text
    df['page']= page_text
    df1 = clean_func(df)
    x_unseen = cv1.transform(df1.clean_text)
    prediction_unseen = load_model.predict(x_unseen, raw_score=True)
    df1['class_pred'] = list(load_model.predict(x_unseen))
    df1['raw_scores'] = list(load_model.predict_proba(x_unseen))
    df1['class_pred'] = df1.class_pred.apply(lambda x: int(x))
    df1 = df1[df1.class_pred== 1]
    if df1.shape[0]>1:
        df1['max_score'] = df1.groupby(['file'])['raw_scores'].transform('max')
        df1['flag'] = np.where(df1['raw_scores'] == df1['max_score'], 1, 0)
        df_t = df1[df1.flag == 1]
   
    list_page=[]
    for i in df_t.page:
        list_page.append(i.split('_page_')[1][0:-4])
    
    return(list_page)
    
```

```python
cv1= pkl.load(open(r'D:\C Drive Backup 2021\Desktop\company_contracts\content_page_classification\version3\cv2.pkl', 'rb'))
load_model = pkl.load(open(r'D:\C Drive Backup 2021\Desktop\company_contracts\content_page_classification\version3\lgbm1.pkl', 'rb'))
path=r'D:\C Drive Backup 2021\Desktop\company_contracts\cgt_data\2020'
outfile= r'D:\C Drive Backup 2021\Desktop\company_contracts\content_page_classification\version3\validation_260_2020_reports.csv'
page_classification_nlp_bulk(path=path, cv1 = cv1, load_model= load_model, outfile=  outfile)
```

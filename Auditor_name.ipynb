```python
def auditor_name_v2(path, file):
    
    
    with fitz.open(os.path.join(path, file)) as f:
        for j in range(f.pageCount):
            page = f.loadPage(j)
            blocks = page.getText("blocks")
            blocks.sort(key=lambda block: block[1])  # sort vertically ascending
            r_text=''
            for b in blocks:
                text = ' '.join(b[4].strip().splitlines())
                r_text=text.lower()
                if ('on behalf' in r_text) and ('statutory auditor' in r_text):
                    pg_num = j
                    return([pg_num, text])
                   
```

```python
def spacy_form(a):
    c={}
    b=[]
    a = eval(str(a))
    for i in a:
        start= i['start']
        end = i['end']
        label=''.join(i['labels'])
        # if label == 'PER':
        #   label = 'PERSON'
        # elif label == 'AREA':
        #   label= 'WORK_OF_ART'
       
        # elif label == 'TITLE':
        #   label = 'NORP'
        b.append((start, end, label))
    c['entities']=b
    return(c)

df= df.dropna()
df['spacy_format'] = df.label.apply(lambda x:spacy_form(x) )
df['train_v1'] =df[['text', 'spacy_format']].apply(lambda x: (x['text'], x['spacy_format']), axis = 1)
spacy_train = [i for i in df.train_v1]
```

```python
# extract results

import cv2
import pytesseract
def transcribed(im):
  image = cv2.imread(im)
  results = model(im)
  #results.print()  
  #results.save()  # or .show()
  results.xyxy[0]  # img1 predictions (tensor)
  a = results.pandas().xyxy[0]
  cc=[]
  for i in range(len(a)):
    y1=int(a['ymin'][i])
    y2= int(a['ymax'][i])
    x1=int(a['xmin'][i])
    x2= int(a['xmax'][i])
    im1= image.copy()
    im2 = im1[y1:y2, x1:x2]
    re_img = cv2.resize(im2, None, fx= 8, fy= 8)
    b = pytesseract.image_to_string(re_img, config = r'--oem 3 --psm 3').strip().splitlines()
    cc.append(b)
  return(cc)
```

```python
################# path to 2020 cgt data
import os
path='/content/gdrive/MyDrive/board_of_directors_segmentation/2021_images_remaining'
file=[]
group=[]
text=[]

for f in os.listdir(path):
  print(f)
  abc = transcribed(os.path.join(path, f))
  if abc:
    for i in range(len(abc)):
      file.append(f)
      group.append(i)
      text.append(abc[i])

```

```python
import pandas as pd
df = pd.DataFrame()
df['filename'] = file
df['group'] = group
df['text'] = text
```

```python
df.head(3)
```

```python
df['clean_text'] = df.text.apply(lambda x: ' '.join(x).strip() )
```

```python
df['clean_text'] = df.clean_text.str.encode('ascii', 'ignore').str.decode('ascii')
```

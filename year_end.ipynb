```python
def year_end_finder(file, path):
       with fitz.open(os.path.join(path, file)) as f:
            for j in range(f.pageCount):
            
                page = f.loadPage(j)
                blocks = page.getText("blocks")
                blocks.sort(key=lambda block: block[1])  # sort vertically ascending
                for b in blocks:
                    text = ' '.join(b[4].strip().splitlines())
                    if 'year ended' in text:
                        val = text.split('year ended')[1][0:20]
                        if ('2020' in val) or ('2021' in val):
                          try:
                            matches = datefinder.find_dates(val)
                            res=[]
                            for match in matches:
                                res.append(match)
                            if res:
                                return(str(res[0]).split(' ')[0])
                          except:
                            continue
```

```python
date_val=[]
file_name=[]
for i in os.listdir(path):
    print(i)
    path = path
    file= i
    date_val.append(year_end_finder(file, path))
    file_name.append(i)
```

```python
df= pd.DataFrame()
df['file_name']= file_name
df['year_end']= date_val
```
